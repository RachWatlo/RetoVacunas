# **Reto - Análisis del progreso mundial de vacunación**

A partir del archivo [country_vaccinations.csv](https://www.kaggle.com/datasets/gpreda/covid-world-vaccination-progress) de Kaggle, se hace lo siguiente.

a. Extraer la información del archivo.



```python
# Importat la biblioteca
import pandas as pd

# Lee el archivo CSV
df = pd.read_csv("/content/country_vaccinations.csv")

# Mostrar las primeras filas del DataFrame para verificar que se haya cargado correctamente
df.head()
```

b. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64.

```python
# Mostrar la estructura y tipos de datos de cada columna
# Todas las columnas parecen ser del tipo correcto, excepto date
df.info()
```

```python
# Se cambia la columna fecha a tipo datetime64
df['date'] = pd.to_datetime(df['date'])

# Las demás columnas de tipo object se convierten a tipo string
# Lista de nombres de columnas por convertir a tipo string
columnas_a_convertir = ['country','iso_code','vaccines','source_name','source_website']

# Utilizando astype para convertir las columnas
df[columnas_a_convertir] = df[columnas_a_convertir].astype('string')

# Mostramos nuevamente la estructura y tipos de datos de cada columna
df.info()
```

c. Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
# Agrupar por vaccines y sumar las vacunas aplicadas
vacunas_por_compania = df.groupby('vaccines')['total_vaccinations'].sum()
vacunas_por_compania
```

```python
# Para verlo en %
# Calcular el total de vacunas aplicadas en todo el mundo
total_vacunas = df['total_vaccinations'].sum()

# Calcular el porcentaje que representa cada compañía respecto al total
porcentaje_por_compania = (vacunas_por_compania / total_vacunas) * 100

# Mostrar los resultados
print(porcentaje_por_compania)
```

```python
# Ordenar los resultados en orden descendente
vacunas_por_compania = vacunas_por_compania.sort_values(ascending=False)
porcentaje_por_compania = porcentaje_por_compania.sort_values(ascending=False)

print(vacunas_por_compania)
print(porcentaje_por_compania)
```

d. Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
# Sumar todas las vacunas aplicadas
total_vacunas = df['total_vaccinations'].sum()
print("Cantidad total de vacunas aplicadas en todo el mundo:", total_vacunas)
```

e. Calcular el promedio de vacunas aplicadas por país.

```python
# Calcular el promedio de vacunas aplicadas por país
promedio_por_pais = df.groupby('country')['total_vaccinations'].mean()
print(promedio_por_pais)
```

```python
# Ordenar los resultados en orden descendente
promedio_por_pais = promedio_por_pais.sort_values(ascending=False)
print(promedio_por_pais)
```

f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
# Filtrar los datos para obtener solo el 29 de enero de 2021 y luego sumar las vacunas aplicadas
vacunas_ene2921 = df[df['date'] == '2021-01-29']['total_vaccinations'].sum()
print("Cantidad de vacunas aplicadas el 29/01/21 en todo el mundo:", vacunas_ene2921)
```

g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.

```python
# Crear una copia del DataFrame original
conDiferencias = df.copy()

# Calcular las diferencias entre las columnas 'daily_vaccinations' y 'daily_vaccinations_raw'
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']

# Mostrar las primeras filas del DataFrame con las diferencias calculadas
conDiferencias.head()
```

h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
# Obtener el periodo de tiempo entre la fecha más reciente y la fecha más antigua
periodo = df['date'].max() - df['date'].min()
print("Periodo de tiempo entre el registro más reciente y el más antiguo:", periodo.days, "días.")
```

i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

```python
# Crear una copia del DataFrame original
conCantidad = df.copy()

# Separar la columna 'vaccines' por el carácter ',' y contar el número de vacunas utilizadas
conCantidad['canVac'] = conCantidad['vaccines'].str.split(',').apply(len)

# Mostrar las primeras filas del DataFrame con la cantidad de vacunas calculada
conCantidad.head()
```

j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.


```python
# Crear un nuevo DataFrame filtrando los registros anteriores al 20 de diciembre de 2020
antes20 = df[df['date'] < '2020-12-20']

# Mostrar las primeras filas del DataFrame antes20
antes20.head()
```

k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
# Filtrar los datos para obtener solo los registros donde se haya utilizado la vacuna Pfizer
pfizer = df[df['vaccines'].str.contains('Pfizer')]
pfizer.head(10)
```

l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.

```python
# Crear un escritor de Excel
with pd.ExcelWriter('/content/resultadosReto.xlsx') as writer:
    # Almacenar cada DataFrame en una hoja diferente
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)
```
